# MEASURES OF CENTRAL TENDENCY (mean, median and mode)

```python
# statistical analysis

# here the dependent or the target variable is "Rain-tomorrow" .Because the other columns are used to predict the whether rains falls tomorrow or not.


# Mean (Average): The sum of all values divided by the total number of values. It represents the central value of the dataset.
# Mean=‚àëùë•ùëñ/ùëõ

# Median: The middle value in an ordered dataset. If there are an odd number of values, it‚Äôs the middle one.
# If even, it‚Äôs the average of the two middle values.

# Mode: The most frequently occurring value in a dataset.
# A dataset can have one mode (unimodal), multiple modes (bimodal or multimodal), or no mode if all values are unique.
```

```python
mean_values = df.mean(numeric_only=True)
print("Mean_Values: \n",mean_values)
```

```python
mode_values = df.mode(numeric_only=True).iloc[0] #here [iloc] is used for selecting data for postion. iloc[0] returns the data of the first row index
print("Mode_Values: \n",mode_values)
```

```python
median_values = df.median(numeric_only=True)
print("Median_Values: \n",median_values)
```

```python
# PRINTS THE SHAPE OF THE DATASET
df.shape
```

```python
# COUNT OF THE DATASET
df.count()
```

```python
# MEDIAN USING GROUPBY METHOD
df.groupby(["MaxTemp","MinTemp"])["MaxTemp"].median().reset_index(name="Median")
```

```python
df.groupby(["MaxTemp","MinTemp"])["MaxTemp"].mean().reset_index(name="Mean")
```

```python
# Here when we give the head as 10 we get a straight line.. which means there is no change in the temperature.
# PLOTING THE MAP/GRAPH
df.groupby(["MaxTemp"])["MinTemp"].count().sort_values(ascending=False).head(10).plot.line(figsize=(14,5))
plt.xlabel("MAX-TEMP")
plt.ylabel("MIN_TEMP")
plt.title("MIN VS MAX", fontsize=20)
plt.show()
```

```python
df.groupby(["MaxTemp"])["MinTemp"].count().sort_values(ascending=False).head(20).plot.line(figsize=(14,5))
plt.xlabel("MAX-TEMP")
plt.ylabel("MIN_TEMP")
plt.legend()
plt.title("MIN VS MAX", fontsize=20)
plt.show()
```

```python
# for finding the corelation for the whole dataset
correlation_matrix=df.corr(numeric_only=True)
correlation_matrix
```

```python
# find the correlation between the max and min temperatures.
corr_matrix = df[["MaxTemp", "MinTemp"]].corr()
corr_matrix
```

```python
# HEATMAP BETWEEN THE MAX AND MIN TEMPERATURE
plt.figure(figsize=(14,5))
sns.heatmap(corr_matrix, annot=True, cmap="Greens" , fmt=".2f")
plt.title("CORRELATION MATRIX b/w MAX&MIN TEMP")
plt.show()
```

```python
# HEATMAP OF THE WHOLE DATASET.

plt.figure(figsize=(14,5))
sns.heatmap(correlation_matrix, annot=True, cmap="Greens" , fmt=".2f")
plt.title("CORRELATION MATRIX of DATASET")
plt.show()
```

```python
from scipy.stats import iqr
print("IQR OF RAINFALL: \n",iqr(df["Rainfall"]))
print("IQR OF MAX-TEMP:\n", iqr(df["MaxTemp"]))
```
